## to_csv

```python
df = pd.read_csv(r"C:\Users\deepu\OneDrive\Desktop\Python\deliveries (1).csv")
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
# Converting the given file into csv and after doing this operation a new file is automatically downloaded
temp_df.to_csv('batsman_runs.csv',index=False)
# A csv file is downloaded inside this folder at home 
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

## to_excel

```python
df
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_excel("batsman_runs.xlsx")
# Dowaloaded a excelfile you can check at home page
```

```python
temp_df2 = df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```
