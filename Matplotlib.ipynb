```python
# Co;ors(hex) and line(width and style) and marker size
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison ")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# Linestyle
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle = 'dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashed')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison ")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# linewith and linestyle
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle = 'dotted',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dotted',linewidth=2)
plt.title("Rohit Sharma vs Virat Kohli Career Comparison ")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

###### Multiple linestyle options sre -->> dashed,dotted,dshdot

```python
# marker at each point and markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linewidth=3,marker='D',markersize=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linewidth=2,marker='D',markersize=2)
plt.title("Rohit Sharma vs Virat Kohli Career Comparison ")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# legend -->> location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linewidth=3,marker='D',markersize=4,label = "Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linewidth=2,marker='D',markersize=2,label = "Rohit Sharma")
plt.title("Rohit Sharma vs Virat Kohli Career Comparison ")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
# Why label is in left side of the curve -->> coz default behaviour is loc='best' means where space is empty then at that
# place it will automatically print
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green greengreen','yellow yellowyellow','pink pinkpink']
plt.bar(colors,children,label = "Favourite Color",color = 'green')
plt.xticks(rotation = 'vertical')
plt.show()
# Here each x-axis name is rotated and visible eaily
```

```python
# stacked Bar chart -->> 
plt.bar(df['batsman'],df['2017'],label = '2017')
plt.bar(df['batsman'],df['2016'],bottom = df['2017'],label = '2016')
plt.bar(df['batsman'],df['2015'],bottom = (df['2016'] + df['2017']),label = '2015')
plt.legend()
plt.title("Total and Season wise runs scored by each Batsman")
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.show()
# By this you are able to know the total runs and season wise runs scored by each batsman
```

#### Pie Chart

##### . Univariate/Bivariate Analysis
##### . Categorical vs Numerical
##### . Use Case - To find contribution on a standard scale

```python
# Simple Example
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels = subjects)
plt.show()
```

```python
# Using dataset
df = pd.read_csv(r"C:\Users\deepu\OneDrive\Desktop\Python\gayle-175.csv")
df
```

```python
plt.pie(df['batsman_runs'],labels = df['batsman'])
plt.show()
```

```python
# Percentage and colors
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = '%0.1f%%',colors = ['blue','green','yellow','red','pink','black'])
plt.show()
```

```python
# Explode shadow -->> for each batsman some value must be assign otherwise error will show(assign 0 if you don't want to explode)
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = "%0.1f%%",explode = [0.1,0,0,0,0,0],shadow = True)
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = "%0.1f%%",explode = [0.1,0,0,0.1,0,0],shadow = True)
plt.show()
```
