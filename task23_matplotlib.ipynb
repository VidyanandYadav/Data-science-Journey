```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# code here
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
df.head()
```

 Problem-1: Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
iran_series = df.query('Country == "Iran"').groupby('Year')['PM2.5'].sum()
china_series = df.query('Country == "China"').groupby('Year')['PM2.5'].sum()
china_series
```

```python
plt.plot(iran_series.index,iran_series.values,linestyle = 'dashed',label = "Iran")
plt.plot(china_series.index,china_series.values,linestyle = 'dashed',label = "China",color = 'red')
plt.title("PM2.5 Over the years")
plt.xlabel("Year")
plt.ylabel("PM2.5")
plt.xticks(df['Year'].value_counts().index)
# NOTE : To show the label you must call legend function otherwise this will not show inside graph
plt.legend()
plt.grid()
plt.show()
```

Problem-2: Draw a histogram of the column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
# Use density = True 
```

```python
plt.hist(df['PM10'],density = True,bins = 50,facecolor = 'green')
plt.xlabel("Bins")
plt.ylabel("Probability")
plt.title("Histogram of PM 10")
plt.grid()
plt.show()
```

Problem-3: Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
poland_data = df.query('Country == "Poland"')
chile_data = df.query('Country == "Chile"')
plt.scatter(poland_data['PM2.5'],poland_data['PM10'],color = 'green',label = "Poland",marker = "D")
plt.scatter(chile_data['PM2.5'],chile_data['PM10'],color = 'red',label = "Chile",marker = "+")
plt.title('PM2.5 and PM10 analysis of two country Poland and Chile')
plt.xlabel("PM2.5")
plt.ylabel('PM10')
plt.legend()
plt.grid()
plt.figure(figsize=(8,6))
plt.show()
```

Problem-4: Draw a pie chart of top 5 most frequent countries.

```python
freque = df['Country'].value_counts().head()
plt.pie(freque,labels=freque.index,autopct="%0.01f%%")
plt.show()
```

Problem-5: Draw a bar chart which represents the counts of top 5 most frequent countries.

```python
freque
```

```python
plt.bar(freque.index,freque,color = 'green',label = "Countrywise Analysis")
plt.title("Number of Top 5 most frequent country")
plt.xlabel("Country")
plt.ylabel("Number of times appear")
plt.legend()
plt.show()
```
