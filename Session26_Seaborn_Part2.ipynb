```python
# swarmplot
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'swarm',hue = 'day')
```

```python
# one more example of hue 
sns.swarmplot(data = tips, x = 'day' , y = 'total_bill', hue = 'sex')
```

#### Categorical Distribution Plots 
##### a. Boxplot
##### b. Violin Plot

#### a.Boxplot
##### A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.
##### Q1 -->> 25th percentile , Q3-->> 75th percentile , Median is the middle of Q1 and Q3 and Min = Q1-1.5*IQR and Max = Q3 + 1.5*IQR and IQR (Interquartile Range which is from Q1 to Q3)

```python
# Box plot -->> axes label func
sns.boxplot(data = tips, x = 'day',y = 'total_bill',hue = 'day')
```

```python
# Using catplot -->> Means using figure label function
sns.catplot(data = tips, x = 'day' , y = 'total_bill', kind = 'box', hue = 'sex')
```

```python
# Single boxplot -->> numerical columns
sns.boxplot(data = tips, y = 'total_bill')
```

##### b.Violinplot = (Boxplot + KDEplot)

```python
# Violinplot
sns.violinplot(data = tips, x = 'day', y = 'total_bill',hue = 'day')
```

```python
# using catplot
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'violin', hue = 'day')
```

```python
# hue parameter
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'violin', hue = 'sex')
```

```python
# Instead of side by side if you want to show both male and female in one diagram use split = True
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'violin', hue = 'sex',split = True)
```

#### Categorical Estimate Plot -> for central tendency
##### a. Barplot
##### b. Pointplot
##### c. Countplot

#### a. Barplot
##### When there are multiple observations in each category , it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars.

```python
import numpy as np
sns.barplot(data = tips, x = 'sex' , y = 'total_bill', hue = 'smoker')
# This gives the average value but you can find any value by passing estimator = np.std/np.mean/np.median
# NOTE : The black line in the bars are called error bars and to remove this use errorbar = None
```

##### NOTE : The black line in the bars are called error bars and to remove this use errorbar = None

```python
sns.barplot(data = tips, x = 'sex' , y = 'total_bill', hue = 'smoker', estimator = np.std, errorbar = None)
```

```python
sns.pointplot(data = tips, x = 'sex' , y = 'total_bill', hue = 'smoker', estimator = np.std , errorbar = None)
```

```python
sns.barplot(data = tips, x = 'sex' , y = 'total_bill', hue = 'smoker', estimator = np.max)
```

```python
sns.pointplot(data = tips, x = 'sex' , y = 'total_bill', hue = 'smoker', estimator = np.max)
```

#### b.Countplot
##### A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical , rather than quantitative variable.

```python
sns.countplot(data = tips , x = 'sex', hue = 'day')
```

#### c .Point Plot
##### Almost Similar to barplot -->> Gives the difference between different quantities

```python
sns.pointplot(data = tips, x = 'sex' , y = 'total_bill', hue = 'smoker',errorbar = None)
```

```python
# faceting using catplot
sns.catplot(data = tips, x = 'sex', y = 'total_bill', col = 'smoker', row = 'time', kind = 'box', hue = 'sex')
```

#### Regression Plots(Regplot)
##### .regplot
##### .lmplot
##### .In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.
#####  Or -->> A regression plot shows the relationship between two variables by fitting a regression line through the data points.

```python
# Let's see a difference -->> first plot scatter plot and then plot a regplot
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

##### regplot

```python
sns.regplot(data = tips, x = 'total_bill', y = 'tip')
# hue parameter is not available
# Here the shadow around the line shows that this much error will do this line
```

##### lmplot -->> It works just like regplot but some minor diff is that you can add hue parameter

```python
# lmplot -->> you can do the same using lmplot (NOTE: Here you can use hue parameter but in regplot it gives error)
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

##### residplot -->> residual plot

```python
sns.residplot(data = tips, x = 'total_bill', y = 'tip')
```

#### Plotting Pairwise Relationship (PairGrid vs Pairplot)

##### PairPlot

```python
iris.head()
```

```python
sns.pairplot(iris)
# Comparison of same quantity on both axis will plot histogram but when you use hue=species then plot KDE plot will plot
```

```python
# same quantity on both axes plots KDE plot automatically
sns.pairplot(iris, hue = 'species')
```

##### NOTE : What is the benefit of PairGrid over PairPlot
##### In Pairplot -->> Offdiag -->> Always scatter plot(you can't make changes in off_diag) and you can change only diagonal plots.
##### In PairGrid -->> Here no any restrictions you can change diag and offdiag plot whatever you want

#### PairGrid

```python
# pairGrid
g = sns.PairGrid(data = iris , hue = 'species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -->> map_offdiag
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# On the basis of diag and offdiag you can plot any two different plot -->> let say diag -> violin and offdiag-->> scatter
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
# If you want diag -->> boxplot and offdiag -->> histplot
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
```

```python
# If you want that diagonal, upper_diag and lower_diag have different plots then it can also be possible
```

```python
# map_diag -->> map_upper -->> map_lower
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# You can restrict that on which columns you want to plot graph by passing vars = [,]
g = sns.PairGrid(data = iris, hue = 'species', vars = ['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

#### JointGrid vs JointPlot

##### JointPlot

######  Available value for `kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']

```python
sns.jointplot(data = tips , x = 'total_bill', y = 'tip',hue = 'sex')
# NOTE : By default kind is scatter
```

```python
sns.jointplot(data = tips , x = 'total_bill', y = 'tip', kind = 'kde',hue = 'sex')
```

```python
sns.jointplot(data = tips , x = 'total_bill', y = 'tip', kind = 'hist', hue = 'sex')
```

```python
sns.jointplot(data = tips , x = 'total_bill', y = 'tip', kind = 'reg')
# NOTE : with kind = 'reg' -->> You cannot use hue parameter
```

##### JointGrid
##### The profit of JointGrid over JointPlot is that you can plot both graph according to your need but in JointPlot you can change only one by passing kind = ''

```python
# Let say we want scatter and histplot which is same as JointPlot but if you want scatter and violin pass this
g = sns.JointGrid(data = tips, x = 'total_bill', y = 'tip')
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data = tips, x = 'total_bill', y = 'tip')
g.plot(sns.kdeplot, sns.violinplot)
```
