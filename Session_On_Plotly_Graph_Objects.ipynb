```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
match = pd.read_csv(r"C:\Users\deepu\OneDrive\Desktop\Python\matches.csv")
delivery = pd.read_csv(r"C:\Users\deepu\OneDrive\Desktop\Python\deliveries (1).csv")
ipl = delivery.merge(match,left_on = 'match_id',right_on = 'id')
ipl.head()
```

```python
# Scatter Plot -->> Strike rate vs batsman run of top 50 batsman
top_50 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(50).index.tolist()
new_ipl = ipl[ipl['batsman'].isin(top_50)]
new_ipl.head()
```

```python
# Calculating SR -->> SR = [(runs scored)/(balls played)]*100
runs = new_ipl.groupby('batsman')['batsman_runs'].sum()
balls = new_ipl.groupby('batsman')['batsman_runs'].count()
sr = (runs/balls)*100
sr = sr.reset_index()
sr.head()
```

```python
# Calculating avg -->> Avg = (runs scored/No of outs) and calculating no of outs(nouts)
out = ipl[ipl['player_dismissed'].isin(top_50)]
nouts = out['player_dismissed'].value_counts()

avg = runs/nouts

avg = avg.reset_index()
avg.rename(columns={'index':'batsman',0:'avg'},inplace = True)

avg = avg.merge(sr,on = 'batsman')
avg.head()
```

```python
# Finally plotting the scatter plot graph
trace = go.Scatter(x = avg['avg'],y = avg['batsman_runs'],
                   mode = 'markers',text = avg['batsman'],
                   marker = {'color':'#00a65a','size':16})

data = [trace]
layout = go.Layout(title = "Batsman Avg vs Strike Rate",
                  xaxis = {'title':'Batsman Runs'},
                  yaxis = {'title':'Batsman Strike Rate'})
fig = go.Figure(data = data , layout = layout)
pyo.plot(fig,filename = 'myfile.html')
```

```python
def batsman_comp(*name):
    data = []
    for i in name:
        single = ipl[ipl['batsman'] == i]
        performance = single.groupby('season')['batsman_runs'].sum().reset_index()

        trace = go.Scatter(x = performance['season'],y = performance['batsman_runs'],
                          mode = 'lines + markers', name = i)
        data.append(trace)
    layout = go.Layout(title = 'Batsman Recoed Comparator',
                      xaxis = {'title':'Season'},
                      yaxis = {'title':'Runs'})
    fig = go.Figure(data = data, layout = layout)

    pyo.plot(fig,filename = 'Year_by_Year.html')
```

```python
batsman_comp('V Kohli','RG Sharma','DA Warner','MS Dhoni')
```

#### Bar Chart
##### Used to show relation between one categorical and 1 numerical data

```python
top10 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).index.tolist()
top10_df = ipl[ipl['batsman'].isin(top10)]
```

```python
top10_score = top10_df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
# Plot Bar Graph
trace = go.Bar(x = top10_score['batsman'], y = top10_score['batsman_runs'])
data = [trace]
layout = go.Layout(title = 'Top 10 IPL Batsman',
                  xaxis = {'title':'Batsman'},
                  yaxis = {'title':'Total Runs'})
fig = go.Figure(data = data , layout = layout)
pyo.plot(fig)
```

#### There are 2 types of Bar Graphs
##### 1.Nested Bar Graph
##### 2.Stacked Bar Graph
##### 3.Overlayed Bar Graph

```python
# Dividing top 10 batsman runs in two part i.e 1st Innings and 2nd Innings
iw = top10_df.groupby(['batsman','inning'])['batsman_runs'].sum().reset_index()
mask1 = iw['inning'] == 1
mask2 = iw['inning'] == 2
one = iw[mask1]
two = iw[mask2]

one.rename(columns={'batsman_runs':'1st Innings'},inplace = True)
two.rename(columns={'batsman_runs':'2nd Innings'},inplace = True)

final = one.merge(two,on='batsman')[['batsman','1st Innings','2nd Innings']]
final
```

```python
# Let's draw all the three types of bar chart one by one
# 1.Overlayed Bar Chart -->> pass barmode = 'overlay'
trace1 = go.Bar(x = final['batsman'], y = final['1st Innings'], name = '1st Innings',
               marker = {'color':'#00a65a'})
trace2 = go.Bar(x = final['batsman'], y = final['2nd Innings'], name = '2nd Innings',
               marker = {'color':'#a6a65a'})

data = [trace1,trace2]

layout = go.Layout(title = 'Inning Wise Scores',
                  xaxis = {'title':'Batsman'},
                  yaxis = {'title':'Runs'},
                  barmode = 'overlay')
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
# Stacked Bar Chart -->> pass barmode = 'stack'
trace1 = go.Bar(x = final['batsman'], y = final['1st Innings'], name = '1st Innings',
               marker = {'color':'#00a65a'})
trace2 = go.Bar(x = final['batsman'], y = final['2nd Innings'], name = '2nd Innings',
               marker = {'color':'#a6a65a'})

data = [trace1,trace2]

layout = go.Layout(title = 'Inning Wise Scores',
                  xaxis = {'title':'Batsman'},
                  yaxis = {'title':'Runs'},
                  barmode = 'stack')
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
# Nested Bar Chart -->> By Default (you don't need to pass barmode = 'something') 
trace1 = go.Bar(x = final['batsman'], y = final['1st Innings'], name = '1st Innings',
               marker = {'color':'#00a65a'})
trace2 = go.Bar(x = final['batsman'], y = final['2nd Innings'], name = '2nd Innings',
               marker = {'color':'#a6a65a'})

data = [trace1,trace2]

layout = go.Layout(title = 'Inning Wise Scores',
                  xaxis = {'title':'Batsman'},
                  yaxis = {'title':'Runs'})
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

#### Bubble Plot
##### Again an extension of Scatter Plot with some additional informations

```python
# In this we use 3 numerical data and 1 categorical data to plot graph let's take a example below
# we have to plot batsman runs scored for each year which is normal because till now we're doing the same thing but additional
# Information is that we can show a large bubble if a batsman hits more sixes and small for that who hit less six(3rd num data)
# And one more is that we can show that from which team batsman belong by some color (this is using categorical information)
```

```python
avg.head()
# Now in this we add one additional columns that total six hits by these batters
new_ipl = new_ipl[new_ipl['batsman_runs'] == 6]
six = new_ipl.groupby('batsman')['batsman_runs'].count().reset_index()
x = avg.merge(six, on = 'batsman')
x.rename(columns = {'batsman_runs_x':'Runs','batsman_runs_y':'Six'},inplace = True)
x.head()
```

```python
# Plotting Bubble Plot
trace = go.Scatter(x = x['avg'], y = x['Runs'],mode = 'markers',
                  marker = {'size':x['Six']},text = x['batsman'])
# NOTE : You can add mode = 'markers+text' -->> So that batsman runs is visible to each bubble although it's visible when 
# You hover on bubbles without doing so.
data = [trace]
layout = go.Layout(title = 'Bubble Chart',
                  xaxis = {'title':'Average'},
                  yaxis = {'title':'Strike Rate'})
fig = go.Figure(data = data , layout=layout)
pyo.plot(fig)
```

#### Histograms

```python
x = delivery.groupby('batsman')['batsman_runs'].count()>150
x = x[x].index.tolist()

new = delivery[delivery['batsman'].isin(x)]

runs = new.groupby('batsman')['batsman_runs'].sum()
balls = new.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100
sr = sr.reset_index()

sr
```

#### Heatmaps
##### A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
six = delivery[delivery['batsman_runs'] == 6]
six = six.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()
six.rename(columns = {'batsman_runs':'No_of_sixes'},inplace = True)
six
```

```python
# Plot Heatmap
trace = go.Heatmap(x = six['batting_team'], y = six['over'], z = six['No_of_sixes'])

data = [trace]

layout = go.Layout(title = 'Six Heatmap')

fig = go.Figure(data=data , layout=layout)

pyo.plot(fig)
```

```python
# Side by Side Heatmap
dots = delivery[delivery['batsman_runs'] == 0]
dots = dots.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()

from plotly import tools

trace1 = go.Heatmap(x = six['batting_team'] , y = six['over'],
                   z = six['No_of_sixes'].values.tolist())

trace2 = go.Heatmap(x = dots['batting_team'] , y = dots['over'],
                   z = dots['batsman_runs'].values.tolist())

fig = tools.make_subplots(rows=1,cols=2,subplot_titles=["6's","0's"],shared_yaxes = True) 


fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)
pyo.plot(fig)
```
