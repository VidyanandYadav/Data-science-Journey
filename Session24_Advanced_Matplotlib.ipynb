#### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx ,yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize = (8,6))
ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
# Another Example
zz = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (8,6))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,zz,cmap = 'viridis')
fig.colorbar(p)
```

```python
# Another one
zzz = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize = (8,6))
ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,zzz,cmap = 'viridis')
fig.colorbar(p)
```

#### Contour Plots 

```python
fig = plt.figure(figsize = (8,6))
ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize = (8,6))
ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

##### Above both graphs are same but only the difference is projection type (2D or 3D) when when see 3D Surface Plots from top it looks like Contour graph which is 2D 

```python
# Filled contour graph
fig = plt.figure(figsize = (8,6))
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
# Let's see above example in Contour 
zz = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (8,6))
ax = plt.subplot()
p = ax.contourf(xx,yy,zz,cmap = 'viridis')
fig.colorbar(p)
```

##### The mistake you're doing in below code is that calling same func again and again as like ax1 = plt.subplot(projection = '3d' and again ax1 = fig.add_subplot(1,2,1,projection = '3d') instead of both this just use ax1 = fig.add_subplot(1,2,1,projection = '3d'))

```python
# Example 1 -->> 
fig = plt.figure(figsize = (12,5))

ax1 = fig.add_subplot(1,2,1,projection = '3d')
# ax1 = plt.subplot(projection = '3d')
ax1.plot_surface(xx,yy,zz,cmap = 'viridis')
ax1.set_title('3D Surface')
ax1.set_xlabel('xx') # Same also for y and z axis

ax2 = fig.add_subplot(1,2,2)
# ax2 = plt.subplot()
ax2.contourf(xx,yy,zz,cmap = 'viridis')
ax2.set_title('2D Contour')

# plt.colorbar(contour,ax = ax2) -->> store plot_surface in a variable contour then call contour and run this to add colorbar
plt.tight_layout()
plt.show()
```

```python
# Example -->> 2
fig = plt.figure(figsize = (12,5))
ax1 = fig.add_subplot(1,2,1,projection = '3d')
# ax1 = plt.subplot(projection = '3d')
ax1.plot_surface(xx,yy,z,cmap = 'viridis')
ax1.set_title('3D Surface')

ax2 = fig.add_subplot(1,2,2)
# ax2 = plt.subplot()
ax2.contourf(xx,yy,z,cmap = 'viridis')
ax2.set_title('2D Contour')

# plt.colorbar(contour,ax = ax2)
plt.tight_layout()
plt.show()
```

#### Heatmap

```python
# We have to plot a such graph where we show how many sixes are hit in every ball of every over from all matches till now
# This will be show using grid and then provide it to heatmap
delivery = pd.read_csv(r"C:\Users\deepu\OneDrive\Desktop\Python\IPL_Ball_by_Ball_2008_2022.csv")
delivery.head()
```

```python
delivery['ballnumber'].unique()
# Means there are some no balls but here we cinsider only 6 balls so first filter this
```

```python
# Considering only 6 balls from each over and batsmanrun == 6 coz we have to do analysis on 6 run at each ball
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]
temp_df.head()
```

```python
# creating pivot table
grid = temp_df.pivot_table(index = 'overs',columns = 'ballnumber',values = 'batsman_run',aggfunc = 'count')
grid.head()
```

```python
# Converting grid line data into graphs
plt.figure(figsize = (15,10))
plt.imshow(grid)
plt.colorbar()
```

```python
# Providing yticks , xticks , title , x_label and y_label 
plt.figure(figsize = (15,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.xlabel('Balls')
plt.ylabel('Overs')
plt.title("Analysis of sixes hitted for every ball of each over ")
plt.colorbar()
```
